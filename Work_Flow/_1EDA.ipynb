```python
# important! check if this method is logically correct or not!

# a function to create permutation test to check if V column is missing at random vs label
def permutation_test(df, test_col, label_col, n=500):
    # a permutation test to check if missing value influence the target value and returns the p value
    # applies a fast permutation test
    
    # the useful df
    df = df[[test_col, label_col]].copy()
    is_null = df[test_col].isnull()
    labels = df[label_col].values
    n_null = is_null.sum()
    n_non_null = len(df) - n_null
    # the target value
    flag = np.sum(is_null * labels)/ n_null \
        - np.sum(~is_null * labels) / n_non_null # calculation: missing perc - nonmissing perc
    
    # make the stock
    is_null_permutations = np.column_stack([
        np.random.permutation(is_null)
        for _ in range(n)
    ]).T
    
    perc_null = (labels * is_null_permutations).sum(axis=1) / n_null
    perc_non_null = (labels * ~is_null_permutations).sum(axis=1) / n_non_null
    ultra_fast = perc_null - perc_non_null
    
    p_value = np.average(ultra_fast >= flag)
    if p_value >= 0.5:
        p_value = 1 - p_value
        
    return p_value
```

```python
%%time
permutation_test(test_df1, 'V1', 'isFraud')
```
